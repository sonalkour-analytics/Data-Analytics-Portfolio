```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
sns.countplot(x="Attrition", data=df)
plt.title("Employee Attrition")
plt.show()
```

```python
sns.countplot(x="Gender", hue="Attrition", data=df)
plt.title("Attrition by Gender")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Department", hue="Attrition", data=df)
plt.xticks(rotation=20)
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(y="JobRole", hue="Attrition", data=df)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["MonthlyIncome"], bins=30, kde=True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], bins=20, kde=True)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x="MaritalStatus", hue="Attrition", data=df)
plt.title("Attrition by Marital Status")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x="OverTime", hue="Attrition", data=df)
plt.title("Attrition by OverTime")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(y="EducationField", hue="Attrition", data=df)
plt.title("Attrition by Education Field")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="BusinessTravel", hue="Attrition", data=df)
plt.title("Attrition by Business Travel")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Attrition", y="Age", data=df)
plt.title("Age vs Attrition")
plt.show()
```

```python
plt.figure(figsize=(18,12))

corr = df.select_dtypes(include=["int64","float64"]).corr()

sns.heatmap(corr, cmap="coolwarm")

plt.title("Correlation Heatmap")

plt.show()
```

```python
attrition = (
    df["Attrition"]
      .value_counts(normalize=True)
      .mul(100)
      .round(2)
)

print(attrition)
```

```python
dept = pd.crosstab(
    df["Department"],
    df["Attrition"],
    normalize="index"
) * 100

print(dept)
```

```python
job = pd.crosstab(
    df["JobRole"],
    df["Attrition"],
    normalize="index"
) * 100

print(job)
```

```python
ot = pd.crosstab(
    df["OverTime"],
    df["Attrition"],
    normalize="index"
) * 100

print(ot)
```
